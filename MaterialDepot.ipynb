```python
from selenium import webdriver
import requests as req
from bs4 import BeautifulSoup as bs
import json
import warnings
warnings.filterwarnings('ignore')
```

```python
driver_path = "C:/Program Files/BraveSoftware/Brave-Browser/Application/chromedriver.exe"
brave_path = "C:/Program Files/BraveSoftware/Brave-Browser/Application/brave.exe"
url='https://materialdepot.in/collections/all'
option = webdriver.ChromeOptions()
option.binary_location = brave_path
# option.add_argument("--incognito") OPTIONAL
# option.add_argument("--headless") OPTIONAL
```

```python
def scraper(url):
    no_of_pages=browser.find_element_by_xpath('//*[@id="shopify-section-collection-template"]/section/div[1]/div[2]/div[2]/div/div/div/div[3]/div/div/a[3]').text
    no_of_pages=int(no_of_pages)
    print(no_of_pages)
    url=url+'?page={i}'
    links=[]
    for i in range(1,no_of_pages+1):
        print("page",i)
        browser.get(url.format(i=i))
        #wait till untill loop completes
        links_page=[]
        for i in range(1,25):
            try:
                element_link=browser.find_element_by_xpath(f'//*[@id="shopify-section-collection-template"]/section/div[1]/div[2]/div[2]/div/div/div/div[2]/div[{i}]/a').get_attribute('href')
                links_page.append(element_link)
            except:
                print('error')
                break
        print(len(links_page))
        links.append(links_page)   
    print(len(links),len(links[0]))
```

```python
# Create new Instance of Chrome
browser = webdriver.Chrome(executable_path=driver_path, chrome_options=option)
browser.get(url)
scraper(url)
browser.close()
```
