```python
# Importing the pandas library, mlxtend.preprocessing, mlxtend.frequent_patterns.
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder 
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import fpgrowth
```

```python
# Creating a dataframe with the columns as the unique values in the dataset.
dataset = [['TFIOS', '5 feet apart', 'Walk to remember'], ['TFIOS', 'Walk to remember'],
['Divergent', 'Harry Potter','Walk to remember'], ['TFIOS', 'Five feet apart'],
['Divergent','Harry Potter']] 
te = TransactionEncoder()
te_array = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_array, columns=te.columns_)
```

```python
df
```

```python
# Using the fpgrowth algorithm to find frequent itemsets. returns support and itemset
frequent_itemsets_fp=fpgrowth(df, min_support=0.02, use_colnames=True)	
```

```python
# A list of frequent itemsets.
frequent_itemsets_fp
```

```python
# Using the frequent itemsets to find rules that have a confidence of at least 80% returns dataframe
rules_fp = association_rules(frequent_itemsets_fp, metric="confidence", min_threshold=0.8)
```
