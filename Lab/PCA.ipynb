## PCA

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("iris.csv")
```

```python
# Import PCA from sklearn
from sklearn.decomposition import PCA
```

```python
df.head()
```

```python
data_PCA = df.iloc[:,1:5]
```

```python
y1 = pd.DataFrame(df["Species"], columns=["Species"])
```

```python
# n_components=2

# Instantiate PCA
pca = PCA(n_components=2)

# Fit PCA to features
principalComponents = pca.fit_transform(data_PCA)

df = pd.DataFrame(data = principalComponents, 
                  columns = ['PC1','PC2'])

target = pd.Series(y1['Species'], name='Species')

result_df = pd.concat([df, target], axis=1)
result_df
```

```python
# n_components=3

# Instantiate PCA
pca = PCA(n_components=3)

# Fit PCA to features
principalComponents = pca.fit_transform(data_PCA)

df = pd.DataFrame(data = principalComponents, 
                  columns = ['PC1','PC2','PC3'])

target = pd.Series(y1['Species'], name='Species')

result_df = pd.concat([df, target], axis=1)
result_df
```
