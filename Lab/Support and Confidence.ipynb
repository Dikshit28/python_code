```python
#Dikshit 19csu379
# Importing the numpy and pandas libraries and giving them an alias.
import numpy as np
import pandas as pd
```

```python
# Reading the csv file and storing it in a dataframe.
df = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
df
```

```python
# Creating a new variable called `X` that is equal to the first column of `df`.
X = df.iloc[:, 0]
X
```

```python
# items is a list of all the items in X.
items=[]
x = list(X)
for i in x:
    for j in i.split(","):
        items.append(j)
```

```python
items
```

```python
# Creating a set of unique items.
item = set(items)
item
len(item)
```

```python
#This is a loop to calculate the support of each item in item.
for j in item: 
    count=0
    for i in x: 
        if j  in i:
            count=count+1     
    support = count/len(x)
    print(j, ":",count , ";support: ", support)
```

```python
# Calculating the confidence of each itemset.
for i in item:
    for j in item:
        count=0
        for k in x:
            # Checking if the itemset i and j are both in the transaction k and if they are not the same item.
            if (i in k) and (j in k) and (i != j):
                count = count+1
        conf = count/len(x)  
        print(i, "=>" ,j, "-", conf)
```
