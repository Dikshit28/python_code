```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('./Position_Salaries.csv')
df
```

```python
x=df.iloc[:,1:2].values
y=df.iloc[:,2].values
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x)
poly_reg.fit(x_poly, y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y)
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg.predict(x), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg_2.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg.predict(x), color = 'green')
plt.plot(x, lin_reg_2.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.title('Polynomial vs Simple Linear Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
def degree_polynomial(degree):
    poly_reg = PolynomialFeatures(degree = degree)
    x_poly = poly_reg.fit_transform(x)
    poly_reg.fit(x_poly, y)
    lin_reg_2 = LinearRegression()
    lin_reg_2.fit(x_poly, y)
    print(f"Poly reg with degree {degree} prediction",lin_reg_2.predict(poly_reg.fit_transform([[6.5]])))

    plt.scatter(x, y, color = 'red')
    plt.plot(x, lin_reg_2.predict(poly_reg.fit_transform(x)), color = 'blue')
    plt.title('Truth or Bluff (Polynomial Regression)')
    plt.xlabel('Position level')
    plt.ylabel('Salary')
    plt.show()
```

```python
for i in range(1,10):
    print("Degree",i)
    degree_polynomial(i)
```

## Conclusion
- Simple Linear Regression is just polyomial regression with degree 1
- Polynomial regression is a generalization of linear regression
- For degree 4 and above the model may not be very accurate for test data (due to overfitting)
- Keeping the degree of polynomial regression low will help in model accuracy

```python
print("Linear reg prediction",lin_reg.predict([[6.5]]))
print("Poly reg with degree 4 prediction",lin_reg_2.predict(poly_reg.fit_transform([[6.5]])))
```
