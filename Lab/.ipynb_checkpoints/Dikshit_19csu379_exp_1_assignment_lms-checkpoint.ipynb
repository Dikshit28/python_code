```python
# dealing with missing data

from sklearn.impute import SimpleImputer 
imputer = SimpleImputer(missing_values = np.nan , strategy ='most_frequent')

# you can replace numerical value by mean, median
# and categorial data can be replaced by mode

imputer = imputer.fit(x.iloc[: , 1:])
x.iloc[:,1:] = imputer.transform(x.iloc[: ,1:])
```

```python
x
```

```python
# therefore in the above table the missing value has been filled with most_frequent of the respective column data.
```

```python
# dealing with label encoding and onehote

from sklearn.preprocessing import LabelEncoder , OneHotEncoder
l = LabelEncoder()
x.iloc[: , 0] = l.fit_transform(x.iloc[: , 0])
```
