```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Generating 200 samples with 2 features, 3 centers, cluster stardard deviation of 1 
data = make_blobs(n_samples=500, n_features=2, centers=5, cluster_std=2, random_state=1)
```

```python
# Plotting the data points with the color corresponding to the class color 'virdis'
plt.figure(figsize=(16,8))
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='viridis')
plt.show()
```

```python
print(data[1])
```

```python
print(data[0])
```

```python
#import kmeans
from sklearn.cluster import KMeans
```

```python
# creating model with 3 clusters
kmeans = KMeans(n_clusters=3)
```

```python
#fitting the model on feature set
kmeans.fit(data[0])
kmeans.cluster_centers_
```

```python
identified_clusters = kmeans.fit_predict(data[0])
identified_clusters
```

```python
#plot of the data points with the color corresponding to the identified cluster
plt.figure(figsize=(14,7))
plt.scatter(data[0][:,0], data[0][:,1], c=identified_clusters, cmap='rainbow')
```

```python
#plot of the centroids and the corresponding cluster
plt.figure(figsize=(14,7))
plt.scatter(data[0][:,0], data[0][:,1], c=identified_clusters, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='black', marker='x')
```

```python
#subplots of original data and clustered data
plt.figure(figsize=(25,7))
plt.subplot(1,2,1)
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='prism')
plt.legend(data[1])
plt.subplot(1,2,2)
plt.scatter(data[0][:,0], data[0][:,1], c=identified_clusters, cmap='prism')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='black', marker='x')
```
