```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Generating 200 samples with 2 features, 3 centers, cluster stardard deviation of 1 
data = make_blobs(n_samples=200, n_features=2, centers=3, cluster_std=1, random_state=1)
```

```python
# Plotting the data points with the color corresponding to the class.
plt.figure(figsize=(14, 7))
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
print("Labels ->",data[1])
```

```python
print(data[0])
```

```python
#import kmeans
from sklearn.cluster import KMeans
```

```python
# creating model with 3 clusters
kmeans = KMeans(n_clusters=3)
```

```python
# fitting the model on feature set
kmeans.fit(data[0])
kmeans.cluster_centers_
```

```python
# Predicting labels for the feature set
identified_clusters = kmeans.fit_predict(data[0])
identified_clusters
```

```python
#plot of the data points with the color corresponding to the identified cluster
plt.figure(figsize=(14,7))
plt.scatter(data[0][:,0], data[0][:,1], c=identified_clusters, cmap='rainbow')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.show()
```

```python
# plot of the centroids and the corresponding cluster
plt.figure(figsize=(14,7))
plt.scatter(data[0][:,0], data[0][:,1], c=identified_clusters, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='black', marker='x')
```

```python
# We use the KMeans class to cluster the data, and then use the identified clusters to assign colors
# to the data points.
# subplots of original data and clustered data
plt.figure(figsize=(25,7))
plt.subplot(1,2,1)
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='black', marker='x')
plt.title('Original Data')
plt.subplot(1,2,2)
plt.scatter(data[0][:,0], data[0][:,1], c=identified_clusters, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='black', marker='x')
plt.title('Clustered Data')
```
