```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Generating 200 samples with 2 features, 3 centers, cluster stardard deviation of 1 
data = make_blobs(n_samples=200, n_features=2, centers=3, cluster_std=2, random_state=1)
```

```python
# Plotting the data points with the color corresponding to the class.
plt.figure(figsize=(14, 7))
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
print("Labels ->",data[1])
```

```python
print(data[0])
```

```python
#import kmeans
from sklearn.cluster import KMeans
```

```python
# creating model with 3 clusters
kmeans = KMeans(n_clusters=3)
```

```python
# fitting the model on feature set
kmeans.fit(data[0])
kmeans.cluster_centers_
```

```python
# Predicting labels for the feature set
identified_clusters = kmeans.fit_predict(data[0])
identified_clusters
```

```python
#plot of the data points with the color corresponding to the identified cluster
plt.figure(figsize=(14,7))
plt.scatter(data[0][:,0], data[0][:,1], c=identified_clusters, cmap='rainbow')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.show()
```

```python
# plot of the centroids and the corresponding cluster
plt.figure(figsize=(14,7))
plt.scatter(data[0][:,0], data[0][:,1], c=identified_clusters, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='black', marker='x')
```

```python
# We use the KMeans class to cluster the data, and then use the identified clusters to assign colors
# to the data points.
# subplots of original data and clustered data
plt.figure(figsize=(25,10))
plt.subplot(1,2,1)
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='black', marker='x')
plt.title('Original Data')
plt.subplot(1,2,2)
plt.scatter(data[0][:,0], data[0][:,1], c=identified_clusters, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='black', marker='x')
plt.title('Clustered Data')
```

```python
from sklearn.metrics.pairwise import euclidean_distances
```

```python
# Computing the distance between each of the cluster centers.
dists = euclidean_distances(kmeans.cluster_centers_)
print(dists)
```

```python
import numpy as np
tri_dists = dists[np.triu_indices(3, 1)]
#upper triangular matrix of distances
print(tri_dists)
max_dist, avg_dist, min_dist = tri_dists.max(), tri_dists.mean(), tri_dists.min()
print("Max distance:", max_dist)
print("Average distance:", avg_dist)
print("Min distance:", min_dist)
# The max, mean, and min distances are calculated from the upper triangle of the distance matrix.
```

```python
wcss = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(data[0])
    wcss.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, wcss, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
import numpy as np
import math
cluster1 = []
cluster2 = []
cluster3 = []


for i in range(0, 3):
    print("inter cluster distance of points from cluster", i+1)
    for j in range(0,200):
        if  kmeans.labels_[j] == i:
            distance = math.dist(kmeans.cluster_centers_[i], data[0][j])
            if i==0:
                cluster1.append(distance)
            elif i==1:
                cluster2.append(distance)
            else:
                cluster3.append(distance)
```

```python
print("cluster 1:", cluster1)
print("cluster 2:", cluster2)
print("cluster 3:", cluster3)
```
