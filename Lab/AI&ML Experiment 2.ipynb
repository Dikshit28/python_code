```python
# Create scaler
scaler = preprocessing.StandardScaler()

# Transform the feature
x = scaler.fit_transform(dataset.iloc[:,:-1])
x=pd.DataFrame(x,columns=dataset.columns[:-1])
x
```

```python
def detect_outliers_zscore(data):
    outliers = []
    mean = round(np.mean(data))
    std = round(np.std(data))
    thres = 3
    for i in data:
        z_score = np.abs((i-mean)/std)
        if (z_score) > thres:
            outliers.append(i)
    return outliers
```

```python
sample_outliers=[]
for i in x.columns:
    sample_outliers.append(detect_outliers_zscore(x.get(i)))

for i in range(len(sample_outliers)):
    print(sample_outliers[i])
    print()
```

```python
def detect_outliers_iqr(data):
    outliers = []
    data = sorted(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    # print(q1, q3)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    # print(lwr_bound, upr_bound)
    for i in data: 
        if (i<lwr_bound or i>upr_bound):
            outliers.append(i)
    return outliers
```

```python
sample_outliers=[]
for i in dataset.columns[:-1]:
    sample_outliers.append(detect_outliers_iqr(dataset.get(i)))

for i in range(len(sample_outliers)):
    print(sample_outliers[i])
    print()
```
