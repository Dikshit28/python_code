## Normal Distribution

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
norm.pdf(x=1.0, loc=0, scale=1)
```

```python
fig, ax = plt.subplots()
x= np.arange(-4,4,0.001)
ax.set_title('N(0,$1^2$)')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.plot(x, norm.pdf(x))
ax.set_ylim(0,0.45)
plt.show()
```

## Normal Dist with different SDVs

```python
fig, ax = plt.subplots()
x = np.linspace(-10,10,100)
std = [1.0, 2.0, 3.0, 4.0]
for s in std:
    #ax.plot(x, norm.pdf(x,scale=s), label='stdv=%.1f' % s)
    ax.plot(x, norm.pdf(x,scale=s, loc=0), label='std=%.1f'%s)
    
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.set_title('Normal Distribution')
#ax.legend(loc='best')
plt.legend()
ax.set_ylim(0,0.45)
#ax.grid(True)
```

## Normal Dist with different means

```python
fig, ax = plt.subplots()
x = np.linspace(-10,20,100)
means = [0.0, 2.0, 4.0, 10.0]
for mean in means:
    ax.plot(x, norm.pdf(x,loc=mean, scale=1), label='mean=%.1f' % mean)
    
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.set_title('Normal Distribution')
ax.legend(loc='best')
ax.set_ylim(0,0.45)
ax.grid(True)
```

## Generating Random Normal Dist Numbers

```python
#norm.rvs(loc=0,scale=1, size=10,  random_state=1)
norm.rvs(loc=0,scale=1, size=10, random_state=2)
```

### np.random.seed(1) or random_state=1 fixes the seed value and generates the same numbers for each execution

```python
fig, ax = plt.subplots()
xs = norm.rvs(scale=2,size=1000)
x = np.linspace(-10,10,100)
p = norm.pdf(x,scale=2)
v = np.var(xs)
m = np.mean(xs)
ax = fig.add_subplot(111)
ax.hist(xs, bins=10, alpha=0.5, density=True)
ax.plot(x,p, 'r-', lw=2)
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.set_title(f'mean={m:.2f}, var={v:.2f}')
ax.grid(True)
```

# A cumulative normal distribution function

The cumulative distribution function of a random variable X, evaluated at x, is the probability that X will take a value less than or equal to x.

```python
from scipy.stats import norm
print("Cumulative distribution function: ", norm.cdf(x=2, loc = 3, scale = 2))
```

```python
fig, ax = plt.subplots()
# for distribution curve
x= np.arange(-4,10,0.001)
ax.plot(x, norm.pdf(x,loc=3,scale=2))
ax.set_title("N(3,$2^2$)")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)
# for fill_between
px=np.arange(-4,2,0.01)
ax.set_ylim(0,0.25)
ax.fill_between(px,norm.pdf(px,loc=3,scale=2),alpha=0.5, color='g')
# for text
ax.text(-0.5,0.02,round(norm.cdf(x=2, loc = 3, scale = 2),2))
```

# Percent Point Function
The percent point function (ppf) is the inverse of the cumulative distribution function. For this reason, the percent point function is also commonly referred to as the inverse distribution function. That is, for a distribution function we calculate the probability that the variable is less than or equal to x for a given x. For the percent point function, we start with the probability and compute the corresponding x for the cumulative distribution. 

```python
norm.ppf(0.3085375387259869,loc=3,scale=2)
```
