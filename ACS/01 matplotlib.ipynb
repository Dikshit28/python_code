## Introduction

Matplotlib was created by John Hunter. He created it to try to replicate MatLab's (another programming language) plotting capabilities in Python. So if you happen to be familiar with matlab, matplotlib will feel natural to you.

It is an excellent 2D and 3D graphics library for generating figures. 

http://matplotlib.org/

### Line plot

```python
plt.figure(figsize=(7,5))
x=np.arange(1,5) 
print(x)
y=2*x 
print(y)
plt.plot([1,2,3,4],[2,4,6,8], 'g*--', linewidth=3) # g is green #* is my marker
#plt.plot(x,y, 'g*--', linewidth=1)
plt.xlabel('time in hours', fontsize=20)
plt.ylabel('Sales')
plt.title('Line plot', fontsize=12)
#plt.grid(True)
plt.savefig("lineplot.png", dpi=100) #pDF
#plt.show()
```

```python
X = np.random.randn(10,2) # (rows,col) # n in rand n stands for normal distribution
print(X)
# rand, randint
```

```python
x_axis = X[0:10 , 0] # (rows,col)
y_axis = X[0:10 , 1]
print(x_axis)
print(y_axis)
```

```python
plt.scatter(x_axis,y_axis, marker="*", c="r", s=20); # s=size of the marker
```

### 2 lines in the same plot

```python
x=np.arange(1,11)
y1=x*2
y2=x*3
print (x,y1,y2)
plt.plot(x,y1) # x vs y1
plt.plot(x,y2) # x vs y2
#plt.show()
```

```python
N = 50
# Generate random numbers
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

## Bar Chart

```python
y = [3, 10, 7, 5, 3, 4.5, 6, 8.1] # 8 items in the list
x = range(len(y))
width = 0.5
plt.bar(x, y, width, color="blue") # indianred
#plt.show()
```

### Histogram

```python
from random import sample
list1=range(1,100) #  list (generated using range ft)
data = sample(list1, 10) # random sampling: returns 10 random vaues from the list
plt.hist(data)
```

```python
## Box and whisker plot: to find an outlier in the data
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)] # std = SD

# rectangular box plot
plt.boxplot(data, vert=True, patch_artist=True);  # default= veritical
# sns.boxplot(x=, optional arguments)
```
